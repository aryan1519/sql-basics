### RDBMS

In the form of tables.

The table is also called a relation.

Each row is known as a tuple.
Each column is known as an attibute.

Primary key is the column which may help in uniquely identifying a row.

Primary key has only unique values.

Primary key has no NULL values.

Primary key is part of the parent table.

Foreign key is a column of the child table.
Foreign key is the primary key but the values may be repeated.
Foreign key is used for cross referencing.

Example

In a parent table of student info student_id,name,age

student_id is the primary key.

There can be another table which has student_id and corresponding course_id of course taken

Here student_id becomes the foreign key.

We can now use this foreign key to get info from this table.

##### Updating a table

###### Adding a column/ Adding multiple columns

-

alter table table_name add(Column_name col_datatype);

alter table table_name add(Column1_name col1_datatype,Column2_name col2_datatype);

-

###### Modifying column dataype

-

alter table table_name modify column_name new_datatype;

-

###### Changing column name

alter table table_name change Old_col_name New_col_name datatype;

-

###### Delete a column or multiple columns

alter table table_name drop column columnName;

alter table table_name drop column column1Name ,drop column column2Name;

-

##### Retrieving data

###### Select

select * from table_name;

select col_name from table_name;

select col1_name , col2_name from table_name;

-

##### Loading data

source file_path;

-

##### Getting disctinct entires from a column

select distinct col_name from table_name;

-

##### Getting desired no of entries

select * from table_name limit no_of_rows;

-

select * from table_name limit no_of_rows offset no_of_rows_to_be_skipped;

-

##### Filter result using where clause

select * from table_name where (condition without brackets);

Example

select * from table_name where column_name = value;

Logical operators such as and & or can also be used.

Strings should be in ''.


NULL related example

select * from orders where col_name is NULL;

select * from orders where col_name is not NULL;

-

##### Aggregate functions

###### count

count(*) - gives no. of rows in table including NULL.


count(column_name) - gives no of rows for a column excluding NULL.


count(distinct column_name) - gives no of rows with distinct values for a column excluding NULL.

select count(*) from table_name;

select count(column_name) from table_name;

select count(distinct column_name) from table_name;

-

###### avg,max,min,sum commands

select avg(column_name) from table_name;

select max(column_name) from table_name;

select min(column_name) from table_name;

select sum(column_name) from table_name;
